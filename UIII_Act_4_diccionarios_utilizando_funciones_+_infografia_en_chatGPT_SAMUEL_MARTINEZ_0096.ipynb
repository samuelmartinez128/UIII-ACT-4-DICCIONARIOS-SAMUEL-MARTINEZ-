#### NL = 35 Fotógrafo: Tipo de cámara, lentes disponibles, especialidad (bodas/moda), estudio, portafolio y resolución de salida

```python
# 1. Inicio con diccionario vacío
Diccionario_fotografo = {}

def listar_fotografo():
    if not Diccionario_fotografo:
        print("\n--- No hay datos registrados ---")
    else:
        # MegaNota: Formato de tabla distribuida
        print("\n" + "-"*110)
        print(f"{'Nombre':<15} | {'Cámara':<15} | {'Lentes':<15} | {'Especialidad':<15} | {'Estudio':<15} | {'Res.'}")
        print("-"*110)

        # Uso de for clave, valor (key, value)
        for nombre, info in Diccionario_fotografo.items():
            # Extraemos los valores del diccionario interno
            c = info['tipo_camara']
            l = info['lentes']
            e = info['especialidad']
            s = info['estudio']
            r = info['resolucion']
            print(f"{nombre:<15} | {c:<15} | {l:<15} | {e:<15} | {s:<15} | {r}")
        print("-"*110 + "\n")

def agregar_fotografo():
    nombre = input("Nombre: ")
    # Guardamos los datos directamente en el diccionario
    Diccionario_fotografo[nombre] = {
        'tipo_camara': input("Tipo de cámara: "),
        'lentes': input("Lentes: "),
        'especialidad': input("Especialidad: "),
        'estudio': input("Estudio: "),
        'resolucion': input("Resolución: ")
    }
    listar_fotografo()

def buscar_fotografo():
    nom = input("Nombre a buscar: ")
    if nom in Diccionario_fotografo:
        print(f"Encontrado: {Diccionario_fotografo[nom]}")
    else:
        print("No existe.")

def borrar_fotografo():
    nom = input("Nombre a eliminar: ")
    if nom in Diccionario_fotografo:
        del Diccionario_fotografo[nom]
    listar_fotografo()

def actualizar_fotografo():
    nom = input("Nombre a actualizar: ")
    if nom in Diccionario_fotografo:
        agregar_fotografo() # Reutilizamos la función para sobreescribir
    else:
        print("No existe.")

# --- Menú interactivo ---
while True:
    print("1. Agregar | 2. Listar | 3. Buscar | 4. Borrar | 5. Actualizar | 6. Salir")
    opcion = input("Selecciona: ")

    if opcion == "1":
        agregar_fotografo()
    elif opcion == "2":
        listar_fotografo()
    elif opcion == "3":
        buscar_fotografo()
    elif opcion == "4":
        borrar_fotografo()
    elif opcion == "5":
        actualizar_fotografo()
    elif opcion == "6":
        break
    else:
        print("Opción no válida")
```
